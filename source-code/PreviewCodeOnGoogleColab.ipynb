```python
import pandas as pd

# 1. Đọc dữ liệu từ file Lazada
df_laz_items = pd.read_csv('20191002-items.csv')
df_laz_reviews = pd.read_csv('20191002-reviews.csv')

# 2. Đọc file Shopee: Thử vách ngăn là dấu chấm phẩy (;) và bỏ qua các dòng lỗi nặng
df_shopee = pd.read_csv('Dataset Review Product di Shopee dan Tokopedia.csv',
                        encoding='latin1',
                        sep=';',
                        on_bad_lines='skip')

# 3. Xem thử 5 dòng đầu tiên của bảng Shopee
df_shopee.head()
```

```python
# --- BƯỚC 1: KIỂM TRA VÀ XÓA DỮ LIỆU TRÙNG LẶP ---

# Đếm xem có bao nhiêu dòng bị trùng lặp hoàn toàn
shopee_trung = df_shopee.duplicated().sum()
print(f"Số dòng trùng lặp trong bảng Shopee: {shopee_trung}")

# Tiến hành xóa các dòng trùng lặp (giữ lại dòng đầu tiên)
df_shopee = df_shopee.drop_duplicates()


# --- BƯỚC 2: KIỂM TRA DỮ LIỆU THIẾU (NULL) ---

# Thao tác này tương tự như việc bạn kiểm tra các trường bị bỏ trống (NULL) trong cơ sở dữ liệu
print("\nSố lượng dữ liệu bị thiếu ở mỗi cột:")
print(df_shopee.isnull().sum())

# Xóa những dòng mà phần nội dung bình luận (review) bị trống
df_shopee = df_shopee.dropna(subset=['review'])

# Kiểm tra lại tổng số dòng hiện còn sau khi làm sạch
print(f"\nTổng số dòng hiện tại của bảng Shopee sau khi làm sạch cơ bản: {len(df_shopee)}")
```

```python
# 1. Loại bỏ các dòng trùng lặp trong 2 file Lazada
df_laz_items = df_laz_items.drop_duplicates()
df_laz_reviews = df_laz_reviews.drop_duplicates()

# 2. Kiểm tra các ô trống (NULL) trong dữ liệu
print("--- SỐ LƯỢNG DỮ LIỆU THIẾU: LAZADA ITEMS ---")
print(df_laz_items.isnull().sum())

print("\n--- SỐ LƯỢNG DỮ LIỆU THIẾU: LAZADA REVIEWS ---")
print(df_laz_reviews.isnull().sum())
```

```python
# --- XỬ LÝ BẢNG LAZADA ITEMS ---
# Điền chữ "Unknown" cho 4 sản phẩm bị thiếu tên thương hiệu
df_laz_items['brandName'] = df_laz_items['brandName'].fillna('Unknown')


# --- XỬ LÝ BẢNG LAZADA REVIEWS ---
# 1. Xóa các dòng không có nội dung đánh giá (xóa dòng null ở cột reviewContent)
df_laz_reviews = df_laz_reviews.dropna(subset=['reviewContent'])

# 2. Điền giá trị mặc định cho các ô trống ở cột tiêu đề và ngày mua
df_laz_reviews['reviewTitle'] = df_laz_reviews['reviewTitle'].fillna('No Title')
df_laz_reviews['boughtDate'] = df_laz_reviews['boughtDate'].fillna('Unknown Date')

# Kiểm tra lại xem bảng Reviews còn bao nhiêu dòng sau khi dọn dẹp
print(f"Tổng số dòng bảng Lazada Reviews sau khi làm sạch: {len(df_laz_reviews)}")
```

```python
# 1. Xuất 3 bảng dữ liệu sạch thành 3 file CSV mới trong hệ thống Colab
# Tham số index=False giúp không in thêm cột số thứ tự thừa vào file
df_laz_items.to_csv('cleaned_lazada_items.csv', index=False)
df_laz_reviews.to_csv('cleaned_lazada_reviews.csv', index=False)

# Riêng Shopee ta lưu lại bằng mã utf-8 để tránh lỗi phông chữ khi nhóm bạn mở bằng Excel
df_shopee.to_csv('cleaned_shopee.csv', index=False, encoding='utf-8')

print("Đã xuất file thành công vào hệ thống Colab! Đang chuẩn bị tải về máy tính...")

# 2. Gọi lệnh của Google Colab để tự động tải 3 file này về máy tính của bạn
from google.colab import files
files.download('cleaned_lazada_items.csv')
files.download('cleaned_lazada_reviews.csv')
files.download('cleaned_shopee.csv')
```
